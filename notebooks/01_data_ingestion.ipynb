```python
# ── Libraries ──────────────────────────────────────────
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Display settings
pd.set_option('display.max_columns', None)  # sare columns dikhenge
pd.set_option('display.float_format', '{:.2f}'.format)  # decimals clean dikhenge

print("✅ Libraries loaded successfully!")
```

```python
## Step 1 — Load All CSV Files
```

```python
# ── File paths define karo ──────────────────────────────
# Tumhara data/raw/ folder yahan hai:
DATA_PATH = 'C:/Users/Sadique/OneDrive/Desktop/ShopRocket-ops-Analytics/data/raw/'

# ── Sare CSV files load karo ek saath ──────────────────
orders       = pd.read_csv(DATA_PATH + 'olist_orders_dataset.csv')
customers    = pd.read_csv(DATA_PATH + 'olist_customers_dataset.csv')
order_items  = pd.read_csv(DATA_PATH + 'olist_order_items_dataset.csv')
payments     = pd.read_csv(DATA_PATH + 'olist_order_payments_dataset.csv')
reviews      = pd.read_csv(DATA_PATH + 'olist_order_reviews_dataset.csv')
products     = pd.read_csv(DATA_PATH + 'olist_products_dataset.csv')
sellers      = pd.read_csv(DATA_PATH + 'olist_sellers_dataset.csv')
geolocation  = pd.read_csv(DATA_PATH + 'olist_geolocation_dataset.csv')
category     = pd.read_csv(DATA_PATH + 'product_category_name_translation.csv')

print("✅ All 9 files loaded successfully!")
```

```python
## Step 2 — Schema Exploration
Har table mein kitni rows, kitne columns hain — aur columns ke naam kya hain
```

```python
# ── Har table ka overview ek saath ─────────────────────
tables = {
    'orders'      : orders,
    'customers'   : customers,
    'order_items' : order_items,
    'payments'    : payments,
    'reviews'     : reviews,
    'products'    : products,
    'sellers'     : sellers,
    'geolocation' : geolocation,
    'category'    : category
}

for name, df in tables.items():
    print(f"{name.upper()}")
    print(f"Rows: {df.shape[0]:,}  |  Columns: {df.shape[1]}")
    print(f"Columns: {list(df.columns)}")
    print()
```

```python
## Step 3 — Data Quality Check
Null values, duplicates aur data types check karna
```

```python
print("=" * 55)
print("NULL VALUES CHECK")
print("=" * 55)

for name, df in tables.items():
    null_count = df.isnull().sum().sum()  # total nulls
    null_pct = (null_count / (df.shape[0] * df.shape[1])) * 100
    status = "Null values" if null_count > 0 else "Clean"
    print(f" {name.upper():15} — {null_count:,} nulls ({null_pct:.1f}%) {status}")
```

```python
### Null values — Column level detail
```

```python
# ── Sirf un tables ka detail dekho jisme nulls hain ────
problem_tables = {
    'orders'   : orders,
    'reviews'  : reviews,
    'products' : products
}

for name, df in problem_tables.items():
    print(f"\n{'='*45}")
    print(f"{name.upper()} — Column-wise Nulls")
    print(f"{'='*45}")
    
    nulls = df.isnull().sum()
    nulls = nulls[nulls > 0]  # sirf wahi columns jo null hain
    
    for col, count in nulls.items():
        pct = (count / len(df)) * 100
        print(f"   {col:45} → {count:,} ({pct:.1f}%)")
```

```python
### Duplicate Records Check
```

```python
# ── Duplicates check ───────────────────────────────────
print("=" * 45)
print("DUPLICATE RECORDS CHECK")
print("=" * 45)

for name, df in tables.items():
    dupes = df.duplicated().sum()
    status = "#" if dupes > 0 else "X"
    print(f"{status} {name.upper():15} — {dupes:,} duplicates")
```

```python
# ── Data Types check — dates strings toh nahi hain? ───
print("\n" + "=" * 45)
print("DATA TYPES — ORDERS TABLE")
print("=" * 45)
print(orders.dtypes)
```

```python
## Step 4 — Data Cleaning
Date columns ko string se datetime mein convert karna
```

```python
# fixing date columns
date_columns = [
    'order_purchase_timestamp',
    'order_approved_at',
    'order_delivered_carrier_date',
    'order_delivered_customer_date',
    'order_estimated_delivery_date'
]

for col in date_columns:
    orders[col] = pd.to_datetime(orders[col], errors='coerce')
    # errors='coerce' menas if not convertible then let it as not a time

# Verification
print("Yes, Date columns converted!")
print("\nDATA TYPES AFTER CONVERSION:")
print(orders[date_columns].dtypes)
```

```python
## Step 5 — Synthetic Enrichment or say feature engineering
Creating column to make it mine personalised - ShopRocket (Indian Marketplace) data
```

```python
import numpy as np

# Seed set karo — taaki har baar same results aayein
# (reproducibility ke liye — interview mein yeh point mention karna!)
np.random.seed(42)

n = len(orders)  # 99,441 rows

# ── 1. Fulfillment Centers ─────────────────────────────
fulfillment_centers = [
    'Mumbai', 'Delhi', 'Bengaluru', 
    'Hyderabad', 'Chennai', 'Kolkata',
    'Pune', 'Ahmedabad', 'Jaipur', 'Lucknow'
]
orders['fulfillment_center'] = np.random.choice(
    fulfillment_centers, 
    size=n, 
    p=[0.20, 0.18, 0.15, 0.10, 0.08, 
       0.08, 0.07, 0.06, 0.04, 0.04]
    # Mumbai aur Delhi zyada orders handle karte hain
)

# ── 2. Carriers ────────────────────────────────────────
carriers = ['Delhivery', 'BlueDart', 'ShadowFax', 'Ecom Express']
orders['carrier'] = np.random.choice(
    carriers,
    size=n,
    p=[0.35, 0.25, 0.25, 0.15]
    # Delhivery sabse bada carrier hai India mein
)

# ── 3. Promised Delivery Days ──────────────────────────
orders['promised_delivery_days'] = np.random.choice(
    [2, 3, 4, 5, 7],
    size=n,
    p=[0.10, 0.25, 0.30, 0.25, 0.10]
)

# ── 4. Actual Delivery Days calculate karo ─────────────
# Real data se nikaalenge — order place se delivery tak
orders['actual_delivery_days'] = (
    orders['order_delivered_customer_date'] - 
    orders['order_purchase_timestamp']
).dt.days

# ── 5. SLA Breach Flag ─────────────────────────────────
# 1 = breach (late delivery), 0 = on time
orders['is_sla_breached'] = np.where(
    orders['actual_delivery_days'] > orders['promised_delivery_days'],
    1, 0
)

# ── 6. Seller Tier ─────────────────────────────────────
seller_tier_map = {}
for seller_id in order_items['seller_id'].unique():
    seller_tier_map[seller_id] = np.random.choice(
        ['Gold', 'Silver', 'Bronze', 'New'],
        p=[0.15, 0.30, 0.35, 0.20]
    )
order_items['seller_tier'] = order_items['seller_id'].map(seller_tier_map)

print("Featured columns added successfully!")
print(f"\n📊 Orders table — New columns:")
print(orders[['fulfillment_center', 'carrier', 
              'promised_delivery_days', 
              'actual_delivery_days',
              'is_sla_breached']].head(10))
```

```python
# ── Overall SLA Breach Rate ────────────────────────────
total_delivered = orders['actual_delivery_days'].notna().sum()
total_breached  = orders['is_sla_breached'].sum()
breach_rate     = (total_breached / total_delivered) * 100

print("=" * 40)
print("📊 SLA BREACH SUMMARY")
print("=" * 40)
print(f"Total Orders    : {len(orders):,}")
print(f"Delivered       : {total_delivered:,}")
print(f"Breached        : {int(total_breached):,}")
print(f"Breach Rate     : {breach_rate:.1f}%")
print(f"On Time         : {100 - breach_rate:.1f}%")
```

```python
# ── Pehle dekho actual delivery days kaisi hain ───────
print("Actual Delivery Days — Distribution:")
print(orders['actual_delivery_days'].describe())
```

```python
# ── Buffer thoda aur badhao ────────────────────────────
np.random.seed(42)

orders['promised_delivery_days'] = np.where(
    orders['actual_delivery_days'].isna(),
    np.random.choice([7, 10, 14], size=len(orders), p=[0.3, 0.4, 0.3]),
    orders['actual_delivery_days'] + np.random.choice(
        [-1, 0, 1, 2, 3, 4, 5, 6, 7],
        size=len(orders),
        p=[0.05, 0.05, 0.10, 0.15, 0.20, 0.20, 0.15, 0.07, 0.03]
    )
)

orders['promised_delivery_days'] = orders['promised_delivery_days'].clip(3, 25)

orders['is_sla_breached'] = np.where(
    orders['actual_delivery_days'].isna(), 0,
    np.where(
        orders['actual_delivery_days'] > orders['promised_delivery_days'],
        1, 0
    )
)

total_delivered = orders['actual_delivery_days'].notna().sum()
total_breached  = orders['is_sla_breached'].sum()
breach_rate     = (total_breached / total_delivered) * 100

print("=" * 40)
print("📊 SLA BREACH SUMMARY (Final)")
print("=" * 40)
print(f"Total Orders    : {len(orders):,}")
print(f"Delivered       : {total_delivered:,}")
print(f"Breached        : {int(total_breached):,}")
print(f"Breach Rate     : {breach_rate:.1f}%")
print(f"On Time         : {100 - breach_rate:.1f}%")
```
